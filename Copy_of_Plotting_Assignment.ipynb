```python
# YOUR CODE HERE
#Set plot size
plt.figure(figsize=(10,6))

#plot histogram
plt.hist(df['co2'], bins=10, edgecolor='black')
#add labes
plt.title('distribution of carbon Dioxide', fontsize=14)
plt.xlabel('CO2 Emissions', fontsize=10)
plt.ylabel('Number of vehicles', fontsize=10)

#show the plot
plt.show()
```

**Carbon Dioxide distribution intepretation**

Based on the histogram above the tallest bar indicates that most cars emit CO2 in the range of approximately 390 to 460

```python
# YOUR CODE HERE
#plot size
plt.figure(figsize=(10, 6))
#plot histogram
plt.hist(df['city'], bins=10, edgecolor='black')

#plot labes
plt.title('City miles per gallon histogram', fontsize=14)
plt.xlabel('City miles per gallon', fontsize=10)
plt.ylabel('Number of vehicles', fontsize=10)
#show plot

plt.show()
```

**City miles per gallon distribution intepretation**

Based on the histogram above the tallest bar indicates that most cars can travel between 16 to 20 miles using one galllon of fuel.

```python
# YOUR CODE HERE
#plot size
plt.figure(figsize=(10, 6))

#plot histogram
plt.hist(df['ghgScore'], bins=10, edgecolor='black')

#plot labes
plt.title('ghg Score', fontsize=14)
plt.xlabel('ghg Score', fontsize=10)
plt.ylabel('Number of vehicles', fontsize=10)
#plot display
plt.show()
```

**ghg score distribution intepretation**

Based on the histogram above the tallest bar indicates that most cars have a ghg score of around 5.

```python
#Select only numbers for correlation
numeric_df = df.select_dtypes(include=['number'])

#make correlation between numeric values
corr_matrix = numeric_df.corr()

#set plot size
plt.figure(figsize=(12, 6))

#Create heatmap
sns.heatmap(corr_matrix, cmap='coolwarm')

#Heatmap label
plt.title('Heatmap of Numerical the fuel econ data', fontsize=14)

#Show
plt.show()
```
